### (2). In a Day Inverstment

We can see that the highest prices is always higher than the opening prices and good for short term investment to consider.

How about in a day inverstment? Previously, we look at the short term and it is the part of short moment. Now, we will see how closing prices will show our profit in a day of conclusion. 

We will use the machine learning to predict the **Close** of closing prices using the independent variable of **Open**, as their correlation coefficient is 0.95 and good to use Linear Regression making variable of closing price prediction as **CP_Predict**

```python
ind3=["Open"]
dep3="Close"

train3 = model.fit(df_nf[ind3], df_nf[dep3])
df_nf["CP_Predict"] = train3.predict(df_nf[ind3])
df_nf
```

We will see the profits in each day of conclusion calculating the price difference of closing and opening prices. The important of the prediction is "Will we make the profit using the prediction in conclusion?", considering if the close prices are higher than opening prices. 

If the Closing prices are higher than the opening, we get the profit. If not, we will lose the money if we are thinking of long term within a day investment.

It is okay either profit or lose. What we want to look at if the prediction model of closing prices will follow similar as actual closing price. Because if the prediction model says it will make profit in conclusion, and what if the actual profit was not true that closing price is lower than opening price in each day. It is bad for investors. If we actually get a profit and the prediction model moves similar way, it is perfect! If it goes opposite, it is bad model for long term within a day prediction.

```python
profits = df_nf["Close"]-df_nf["Open"]
prediction_profits = df_nf["CP_Predict"]-df_nf["Open"]
```

If it shows 1, the prediction is opposite between actual and precition. If it is 0, they moved similar way in each day of stock market.

```python

data=[]

for i in range(len(df_nf)):
    profits = df_nf.loc[i, "Close"]-df_nf.loc[i, "Open"]
    prediction_profits = df_nf.loc[i, "CP_Predict"]-df_nf.loc[i, "Open"]
    if profits >0 and prediction_profits>0:
        error = 0

    if profits <0 and prediction_profits<0:
        error= 0

    if profits >0 and prediction_profits<0:
        error=1

    if profits <0 and prediction_profits>0:
        error=1
    
    d = {"error":error}
    data.append(d)
df_error_profits = pd.DataFrame(data)
df_error_profits
```

Oh no, we can see that there is difference between actual profit and prediction profit. Let's see how frequently prediction model made action differently with actual ones.

```python
df_error_profits_show = df_error_profits[df_error_profits.error > 0]
df_error_profits_show
```
